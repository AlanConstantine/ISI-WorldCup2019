# Features Select(var)

```python
df_n = (df- df.min())/ (df.max() - df.min())
stddf = df_n.describe().loc[['std']]
stddf.drop(['企业编号'],axis=1,inplace=True)
stddf.loc['variance'] = stddf.loc['std']**2
x_max = pd.DataFrame()
x_max['feature'] = stddf.columns
features=stddf.T.sort_values(by=['variance'], ascending=False)
features['rank']=(np.arange(1,len(features)+1))
features.loc[(features['variance']<0.05)==True,'variance']=np.nan
features=features.dropna()
features
```

# Normalize

```python
newdf = df[features.T.columns]
newdf=(newdf-newdf.mean())/newdf.std()
newdf['企业编号'] = df['企业编号']
newdf['企业总评分'] = df['企业总评分']
newdf.to_csv('.././Documents/赛题1数据集/Ntrain_data.csv')
```
