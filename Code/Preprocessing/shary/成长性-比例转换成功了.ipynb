成长能力分析


```python
import pandas as pd
df_growth_bkp = pd.read_excel(r'../../../Data/CompInfo/上市信息财务信息-成长能力指标.xlsx')
df_growth = df_growth_bkp.copy()
df_growth.head(n=5)
A = ['营业总收入(元)', '毛利润(元)', '归属净利润(元)', '扣非净利润(元)']
```

```python
# 剔除单位
for arr in A:
    chinese = set(df_growth[arr].str.extract(
        "([^0-9\.\-]+)", expand=False).dropna())
    print(arr, chinese)
```

```python
for arr in A:
    df_growth.loc[(df_growth[arr].str.contains("万亿") == True), arr] = df_growth.loc[(
        df_growth[arr].str.contains("万亿") == True), arr].str.replace('万亿', '').astype(float)*1000000000000
    df_growth.loc[(df_growth[arr].str.contains("万") == True), arr] = df_growth.loc[(
        df_growth[arr].str.contains("万") == True), arr].str.replace('万', '').astype(float)*10000
    df_growth.loc[(df_growth[arr].str.contains("亿") == True), arr] = df_growth.loc[(
        df_growth[arr].str.contains("亿") == True), arr].str.replace('亿', '').astype(float)*100000000
```

```python
for arr in A:
    chinese = set(df_growth[arr].str.extract(
        "([^0-9\.\-]+)", expand=False).dropna())
    print(arr, chinese)
```

```python
# 增长比例去百分号处理
import numpy as np
B = ['营业总收入同比增长(元)', '归属净利润同比增长(元)', '扣非净利润同比增长(元)',
     '营业总收入滚动环比增长(元)', '归属净利润滚动环比增长(元)', '扣非净利润滚动环比增长(元)']
for arr in B:
    df_growth[arr] = df_growth[arr].replace('--%', np.nan)
df_growth.info()
```

```python
df_growth = df_growth.convert_objects(convert_numeric=True)
```

```python
for arr in B:
    df_growth.loc[(df_growth[arr].str.contains("%") == True), arr] = df_growth.loc[(
        df_growth[arr].str.contains("%") == True), arr].str.replace(',', '').str.rstrip('%').astype(float)/100
```

```python
# 对数值型指标和评分的相关性分析
agg_func = {
    '营业总收入(元)': ['mean', 'max', 'min', 'std'],
    '毛利润(元)': ['mean', 'max', 'min', 'std'],
    '归属净利润(元)': ['mean', 'max', 'min', 'std'],
    '扣非净利润(元)': ['mean', 'max', 'min', 'std'],
}
df_growth_new = df_growth.groupby(['企业编号']).agg(agg_func)
df_growth_new.columns = ['_'.join(col) for col in df_growth_new.columns.values]
df_growth_new.columns = [
    '营业总收入(元)_mean', '营业总收入(元)_max', '营业总收入(元)_min', '营业总收入(元)_std', '毛利润(元)_mean', '毛利润(元)_max', '毛利润(元)_min', '毛利润(元)_std',
    '归属净利润(元)_mean', '归属净利润(元)_max', '归属净利润(元)_min', '归属净利润(元)_std', '扣非净利润(元)_mean', '扣非净利润(元)_max', '扣非净利润(元)_min', '扣非净利润(元)_std']
df_growth_new['企业编号'] = df_growth_new.index
df_growth_new.head(n=3)
```

```python
df_growth_xiangguanfenxi = df_growth_new.copy()
df_growth_xiangguanfenxi.info()
score = pd.read_excel(r'.././Documents/赛题1数据集/企业评分.xlsx')
score.info()
```

```python
df_growth_xiangguanfenxicor = df_growth_xiangguanfenxi.merge(
    score, how='inner', left_index=True, right_on="企业编号")
df_growth_xiangguanfenxicor.info()
columns = ['企业总评分',
           '营业总收入(元)_mean', '营业总收入(元)_max', '营业总收入(元)_min', '营业总收入(元)_std', '毛利润(元)_mean', '毛利润(元)_max', '毛利润(元)_min', '毛利润(元)_std',
           '归属净利润(元)_mean', '归属净利润(元)_max', '归属净利润(元)_min', '归属净利润(元)_std', '扣非净利润(元)_mean', '扣非净利润(元)_max', '扣非净利润(元)_min', '扣非净利润(元)_std']
df_growth_xiangguanfenxicor = pd.DataFrame(
    df_growth_xiangguanfenxicor, columns=columns)
```
