```python
# 利润表特征相关性分析
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib as mpl
from IPython.display import display
from sklearn.preprocessing import Imputer
from scipy import stats
plt.style.use("fivethirtyeight")
sns.set_style({'font.sans-serif':['simhei','Arial']})
profitdataset= pd.read_excel('/Users/zengxiaoyu/Documents/赛题1数据集/上市信息财务信息-利润表(清楚单位).xlsx')
profitdataset=pd.DataFrame(profitdataset)
profitdataset.info()
profitdataset["销售费用(元)"]=pd.to_numeric(profitdataset["销售费用(元)"],errors='coerce')
profitdataset["资产减值损失(元)"]=pd.to_numeric(profitdataset["资产减值损失(元)"], errors='coerce')
profitdataset["投资收益(元)"]=pd.to_numeric(profitdataset["投资收益(元)"], errors='coerce')
profitdataset["所得税(元)"]=pd.to_numeric(profitdataset["所得税(元)"], errors='coerce')
profitdata1=profitdataset.dropna()
profitdata1.info()
#总样本：19621
#完整样本：15973
#比例：81.4%
```
