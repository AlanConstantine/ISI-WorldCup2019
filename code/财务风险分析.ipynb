财务风险分析

```python
import pandas as pd
import numpy as np
df_risk= pd.read_excel(r'.././Documents/赛题1数据集/上市信息财务信息-财务风险指标.xlsx')
df_risk.info()
A=['流动比率','速动比率']
for arr in A:
      df_risk[arr] = pd.to_numeric(df_risk[arr],errors='coerce')
B=['资产负债率(%)','流动负债/总负债(%)']
for arr in B:
     df_risk[arr] = df_risk[arr].replace('--%', np.nan)
df_risk.convert_objects(convert_numeric=True)
for arr in B:
      df_risk[arr] =df_risk[arr].str.rstrip('%').astype(float)/100
df_risk.info()
```

```python
df_risk=df_risk.dropna()
df_risk.corr()
```

```python
agg_func = {
    '资产负债率(%)': ['mean', 'max', 'min','std'],
    '流动负债/总负债(%)': ['mean', 'max', 'min','std'],
    '流动比率': ['mean', 'max', 'min','std'],
    '速动比率': ['mean', 'max', 'min','std'],
}
df_risk_new = df_risk.groupby(['企业编号']).agg(agg_func)
df_risk_new.columns = ['_'.join(col) for col in df_risk_new.columns.values]
df_risk_new.columns = [
     '资产负债率(%)_mean', '资产负债率(%)_max', '资产负债率(%)_min', '资产负债率(%)_std','流动负债/总负债(%)_mean', '流动负债/总负债(%)_max', '流动负债/总负债(%)_min', '流动负债/总负债(%)_std',
     '流动比率_mean', '流动比率_max', '流动比率_min', '流动比率_std', '速动比率_mean', '速动比率_max', '速动比率_min', '速动比率_std']
df_risk_new['企业编号'] =df_risk_new.index
df_risk_new.head(n=3)
df_risk_new.info()
```

```python
score= pd.read_excel(r'.././Documents/赛题1数据集/企业评分.xlsx')
df_risk_cor = df_risk_new.merge(score, how='inner', left_index=True,right_on="企业编号")
columns = ['企业总评分',
           '资产负债率(%)_mean', '资产负债率(%)_max', '资产负债率(%)_min', '资产负债率(%)_std','流动负债/总负债(%)_mean', '流动负债/总负债(%)_max', '流动负债/总负债(%)_min', '流动负债/总负债(%)_std',
     '流动比率_mean', '流动比率_max', '流动比率_min', '流动比率_std', '速动比率_mean', '速动比率_max', '速动比率_min', '速动比率_std']
df_risk_cor = pd.DataFrame(df_risk_cor, columns=columns)
df_risk_cor.info()
df_risk_cor=df_risk_cor.dropna()
df_risk_cor.info()
# 缺失情况 ：2995-2926=69个企业样本 缺失来源是流动比例和速动比例  两
```
