```python
agg_func = {'债券期限': ['mean', 'max', 'min','std'],'票面利率（%）': ['mean', 'max', 'min','std'],'计划发行总额（亿元）': ['sum', 'max', 'min','std']}
df_operation_new = df_operation.groupby(['企业编号']).agg(agg_func)
df_operation_new.columns = ['_'.join(col) for col in df_operation_new.columns.values]
df_operation_new.columns = [
    '债券期限_mean', '债券期限_max', '债券期限_min', '债券期限_std', 
    '票面利率（%）_mean', '票面利率（%）_max', '票面利率（%）_min', '票面利率（%）_std','计划发行总额（亿元）_sum','计划发行总额（亿元）_max','计划发行总额（亿元）_min','计划发行总额（亿元）_std']
df_operation_new['企业编号'] = df_operation_new.index
df_operation_new.head(n=3)
df_operation_new.info()
```

```python
df_operation_xiangguanfenxi=df_operation_new.copy()
df_operation_xiangguanfenxi.info()
score= pd.read_excel(r'.././Documents/赛题1数据集/企业评分.xlsx')
score.info()
```

```python
df_operation_xiangguanfenxicor=df_operation_xiangguanfenxi.merge(score, how='inner', left_index=True,right_on="企业编号")
df_operation_xiangguanfenxicor.info()
columns = ['企业总评分',
       '债券期限_mean', '债券期限_max', '债券期限_min', '债券期限_std', 
    '票面利率（%）_mean', '票面利率（%）_max', '票面利率（%）_min', '票面利率（%）_std','计划发行总额（亿元）_sum','计划发行总额（亿元）_max','计划发行总额（亿元）_min','计划发行总额（亿元）_std']
df_operation_xiangguanfenxicor = pd.DataFrame(df_operation_xiangguanfenxicor, columns=columns)
```

```python
#新特征引入 既平均每元每年支付的利息=利率*期限*发行金额/期限*发行金额   
df_operation['利息合计']=df_operation['债券期限']*df_operation['票面利率（%）']*df_operation['计划发行总额（亿元）']
df_operation['时间金额合计']=df_operation['债券期限']*df_operation['计划发行总额（亿元）']
agg_func = {'利息合计': ['sum'],'时间金额合计': ['sum'],}
df_operation_new1 = df_operation.groupby(['企业编号']).agg(agg_func)
df_operation_new1.columns = ['_'.join(col) for col in df_operation_new1.columns.values]
df_operation_new1.columns = [
    '利息合计_sum', '时间金额合计_sum']
df_operation_new1['企业编号'] = df_operation_new1.index
df_operation_new1.head(n=3)
df_operation_new1.info()
```

```python
df_operation_new1["平均利息"]=df_operation_new1["利息合计_sum"]/df_operation_new1["时间金额合计_sum"]
df_operation_xiangguanfenxicor1=df_operation_new1.merge(score, how='inner', left_index=True,right_on="企业编号")
df_operation_xiangguanfenxicor1.info()
columns = ['企业总评分','平均利息']
df_operation_xiangguanfenxicor1 = pd.DataFrame(df_operation_xiangguanfenxicor1, columns=columns)
```
