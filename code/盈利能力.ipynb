```python
import pandas as pd
import numpy as np
df_abilityofprofit= pd.read_excel(r'.././Documents/赛题1数据集/上市信息财务信息盈利能力指标.xlsx')
```

```python
A=['加权净资产收益率(%)','摊薄净资产收益率(%)','摊薄总资产收益率(%)','毛利率(%)','净利率(%)','实际税率(%)']
for arr in A:
      df_abilityofprofit[arr] = df_abilityofprofit[arr].replace('--%', np.nan)
df_abilityofprofit.info()
df_abilityofprofit.convert_objects(convert_numeric=True)
A=['加权净资产收益率(%)','摊薄净资产收益率(%)','摊薄总资产收益率(%)','毛利率(%)','净利率(%)','实际税率(%)']
for arr in A:
      df_abilityofprofit[arr] =df_abilityofprofit[arr].str.rstrip('%').astype(float)/100
df_abilityofprofit.info()
df_abilityofprofit.corr()
```

```python
df_abilityofprofit_del=df_abilityofprofit.dropna()
df_abilityofprofit_del.info()
df_abilityofprofit_del.corr()
```

```python
agg_func = {
    '加权净资产收益率(%)': ['mean', 'max', 'min','std'],
    '摊薄净资产收益率(%)': ['mean', 'max', 'min','std'],
    '摊薄总资产收益率(%)': ['mean', 'max', 'min','std'],
    '毛利率(%)': ['mean', 'max', 'min','std'],
    '净利率(%)':['mean', 'max', 'min','std'],
}
df_abilityofprofit_new = df_abilityofprofit.groupby(['企业编号']).agg(agg_func)
df_abilityofprofit_new.columns = ['_'.join(col) for col in df_abilityofprofit_new.columns.values]
df_abilityofprofit_new.columns = [
     '加权净资产收益率(%)_mean', '加权净资产收益率(%)_max', '加权净资产收益率(%)_min', '加权净资产收益率(%)_std','摊薄净资产收益率(%)_mean', '摊薄净资产收益率(%)_max', '摊薄净资产收益率(%)_min', '摊薄净资产收益率(%)_std',
     '摊薄总资产收益率(%)_mean', '摊薄总资产收益率(%)_max', '摊薄总资产收益率(%)_min', '摊薄总资产收益率(%)_std', '毛利率(%)_mean', '毛利率(%)_max', '毛利率(%)_min', '毛利率(%)_std',
     '净利率(%)_mean', '净利率(%)_max', '净利率(%)_min', '净利率(%)_std']
df_abilityofprofit_new['企业编号'] =df_abilityofprofit_new.index
df_abilityofprofit_new.head(n=3)
```

```python
df_abilityofprofit_new.info()
```

```python
score= pd.read_excel(r'.././Documents/赛题1数据集/企业评分.xlsx')
score.info()
```

```python
df_abilityofprofit_cor = df_abilityofprofit_new.merge(score, how='inner', left_index=True,right_on="企业编号")
df_abilityofprofit_cor.info()
columns = ['企业总评分',
              '加权净资产收益率(%)_mean', '加权净资产收益率(%)_max', '加权净资产收益率(%)_min', '加权净资产收益率(%)_std','摊薄净资产收益率(%)_mean', '摊薄净资产收益率(%)_max', '摊薄净资产收益率(%)_min', '摊薄净资产收益率(%)_std',
     '摊薄总资产收益率(%)_mean', '摊薄总资产收益率(%)_max', '摊薄总资产收益率(%)_min', '摊薄总资产收益率(%)_std', '毛利率(%)_mean', '毛利率(%)_max', '毛利率(%)_min', '毛利率(%)_std',
     '净利率(%)_mean', '净利率(%)_max', '净利率(%)_min', '净利率(%)_std']
df_abilityofprofit_cor = pd.DataFrame(df_abilityofprofit_cor, columns=columns)
```

```python
df_abilityofprofit_cor=df_abilityofprofit_cor.dropna()
df_abilityofprofit_cor.info()
df_abilityofprofit_cor.corr()
# 特征缺失数2956-2924=32
```
